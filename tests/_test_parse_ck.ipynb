```python
def ckHtmlToChaps(fn):
    bn = os.path.splitext(os.path.basename(fn))[0]
    with open(fn, 'rt') as fp:
        html = fp.read()
        d = pq(html)
    # d = pq(html)
    articles = d('td[class="article_plc_user"]')('table:first')('td:first')
    page_title = d('td[class="article_plc_user"]:first')('strong:first').text()
    if page_title == "":
        page_title = d('td[class="article_plc_user"]:first')('table:first')('td:first').text().split('\n')[0]
    
    print(len(articles), page_title)
    chaps = []
    i = 0
    for a in articles:
        page_title = pq(a)('strong:first').text()
        if page_title == "":
            page_title = a.text.strip().split('\n')[0]
        print(page_title)
        b = deepcopy(a)
        b.tag = "div"
        s = etree.tostring(b, encoding='UTF-8').decode('utf-8')
        c = epub.EpubHtml(
            title=page_title,
            file_name=f'{bn}-{i+1}.xhtml',
            lang='zh-Hant'
        )
        c.set_content(s)
        chaps.append(c)
        i += 1
    return chaps
# rs += "</body></html>\r\n"
# print(rs)
# rs = "<html><head></head><body>\r\n"
ckHtmlToChaps('input/p78.html')
ckHtmlToChaps('input/p105.html')
```

```python
def ckParseHTML(fn):
    with open(fn, 'rt') as fp:
        html = fp.read()
        d = pq(html)
    # d = pq(html)
    rs = ""
    articles = d('td[class="article_plc_user"]')('table:first')('td:first')
    page_title = d('td[class="article_plc_user"]:first')('strong:first').text()
    if page_title == "":
        page_title = d('td[class="article_plc_user"]:first')('table:first')('td:first').text().split('\n')[0]
    
    print(len(articles), page_title)
    for a in articles:
        b = deepcopy(a)
        b.tag = "div"
        s = etree.tostring(b, encoding='UTF-8').decode('utf-8')
        rs += s
    return page_title, rs
# rs += "</body></html>\r\n"
# print(rs)
# rs = "<html><head></head><body>\r\n"
page_title, rs = ckParseHTML('input/p105.html')
page_title, rs
```
